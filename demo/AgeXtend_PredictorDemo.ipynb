```python
#Importing Predictior libraries 
from AgeXtend import Predictor
```

```python
#Preparing list of query compound 
smiles = ['C(CCNCCCN)CNCCCN', 'C(CCNCCCN)CN'] ## Known Anti Aging Compounds: spermine, spermidine, respectively
```

```python
#Generating features for query compounds
agex_obj = Predictor.featurize(smiles)
```

```python
#Generating prediction for query compound 
output = Predictor.predict(agex_obj)
```

```python
#Analyzing the AgeXtend output object
output.keys()
```

```python
#Getting AgeXtend sub-module specific outputs
output['Explainability_Status']
```
